5.1 Выберите дом клиенту
Клиент хочет дом с видом на набережную, как минимум с тремя ванными и с подвалом. \
Сколько вариантов есть у клиента?

```python
x1 = df[(df['waterfront'] == 1) & (df['bathrooms'] >= 3)& (df['sqft_basement'] != 0)]
x1.shape[0]
```

5.2 Выберите дом клиенту
Клиент хочет дом либо с очень красивым видом из окна, либо с видом на набережную, в очень хорошем состоянии и год постройки не меньше 1980 года. В какой ценовом диапазоне будут дома?

```python
x2 = (df['view'] <= 3) | (df['waterfront'] == 1) & (df['condition'] <= 4) & (df['yr_built'] >= 1980)
round(df[x2]['price'].min()),round(df[x2]['price'].max())
```

5.3 Выберите дом клиенту

```python
x3 = df[(df['sqft_basement'] == 0) & (df['floors'] == 2)& (df['price'] < 150000)]
by_condition = round(x3['condition'].mean(),2)
by_grade = round(x3['grade'].mean())
by_condition, by_grade
```
